```python
# List of columns for Y-axes
y_columns = ['breastcancerper100th', 'internetuserate', 'lifeexpectancy', 'relectricperperson', 'urbanrate']

# Set the style of the plots
sns.set(style="whitegrid")

# Create subplots with 3 rows and 2 columns
fig, axes = plt.subplots(nrows=math.ceil(len(y_columns)/2), ncols=2, figsize=(12, 15))

# Flatten the axes for iteration
axes = axes.flatten()

# Iterate through the columns and create scatter plots
for i, y_column in enumerate(y_columns):
    if i < len(axes):
        ax = axes[i]
        sns.scatterplot(x='region', y=y_column, data=bivariate_analysis_df, hue='region', palette='Set2', s=100, ax=ax)
        ax.set_xlabel('Region')
        ax.set_ylabel(y_column)  # Set Y-axis label
        ax.set_title(f'{y_column} by Region')
        ax.set_xticks(range(len(bivariate_analysis_df['region'].unique())))  # Set tick positions
        ax.set_xticklabels(bivariate_analysis_df['region'].unique(), rotation=90)  # Set tick labels
        ax.get_legend().remove()  # Remove the legend for each subplot

# Remove the last empty subplot
if len(y_columns) % 2 != 0:
    fig.delaxes(axes[-1])

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

Vemos a partir de el análisis por región que sin duda existe una correlación entre la mayoría de variables numéricas y la variable categórica incluida (región) lo que indica indirectamente que se espera también una correlación significativa entre la región y la variabla de respuesta.

## [30%] Entrenamiento del modelo de regresión
En esta etapa se realizará la preparación, entrenamiento, verificación de supuestos y interpretación del modelo de regresion

### Preparación
En esta etapa se eliminarán las columnas no necesarias para el análisis (como el identificador de país) se realizará el encoding de las variables categóricas, se separará el dataset en 'x' y 'Y' y se separará en datos de entrenamiento y de prueba.

Empezamos eliminando el identificador de país

```python
final_df.drop(columns=['country'], inplace=True)
```

Como ambas variables categóricas que tenemos (region y income) son nominales, se aplicará OneHotEncoder para su codificación a variables numéricas para poder aplicar el modelo de regresión

```python
encoded_columns = pd.get_dummies(final_df, columns=['region', 'income'], drop_first=True)
encoded_columns = encoded_columns.astype(float).drop(['incomeperperson', 'breastcancerper100th', 'internetuserate', 'lifeexpectancy', 'relectricperperson', 'urbanrate'], axis=1)
final_df_encoded = pd.concat([final_df.drop(['region', 'income'], axis=1), encoded_columns], axis=1)
```

```python
final_df_encoded.info()
```

```python
final_df_encoded.sample(10)
```

Obtenemos los nombres de las features

```python
feature_names = final_df_encoded.drop('incomeperperson', axis=1).columns.to_numpy()
print("Feature Names:", feature_names)
```

A partir de la información de las columnas y de la muestra de los datos, vemos que se encuentran listos para ser divididos en 'X' y 'Y'

```python
X = final_df_encoded.drop(columns=['incomeperperson'])
Y = final_df_encoded['incomeperperson']
```

Dividimos los datos en conjunto de entrenamiento y de prueba (80%-20%)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(25, 6))
ax1.boxplot(Y_train, showmeans=True, vert=False)
ax2.boxplot(Y_test, showmeans=True, vert=False)
plt.show()
```

Normalizamos las features utilizando StandardScaler

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Inicializamos el modelo de regresión

```python
model = LinearRegression()
```

Entrenamos el modelo

## Interpretación del modelo

```python
preds_train = model.predict(X_train_scaled)
preds_test = model.predict(X_test_scaled)
```

```python
print("MAE train:", mean_absolute_error(Y_train, preds_train))
print("MAE test:", mean_absolute_error(Y_test, preds_test))
```

```python
print("RMSE train:", np.sqrt(mean_squared_error(Y_train, preds_train)))
print("RMSE test:", np.sqrt(mean_squared_error(Y_test, preds_test)))
```

```python
print("R2 score:", r2_score(Y_train, preds_train))
```

El modelo de regresión desarrollado presenta un buen rendimiento en la predicción del ingreso per cápita, con métricas de evaluación robustas tanto en el conjunto de entrenamiento como en el conjunto de prueba. El error absoluto medio (MAE) en el conjunto de entrenamiento es de aproximadamente 2676.24 unidades monetarias, lo que significa que, en promedio, las predicciones del modelo difieren en esta cantidad del valor real del ingreso per cápita. En el conjunto de prueba, el MAE es ligeramente mayor, con un valor de alrededor de 2808.17 unidades monetarias, indicando una buena generalización del modelo a datos no vistos. Además, el error cuadrático medio (RMSE) en el conjunto de entrenamiento es de aproximadamente 4202.19 unidades monetarias, y en el conjunto de prueba es de alrededor de 4394.59 unidades monetarias, lo que demuestra la capacidad del modelo para capturar la variabilidad en los datos de manera efectiva. Además, el coeficiente de determinación (R²) es 0.8421, lo que indica que aproximadamente el 84.21% de la variación en el ingreso per cápita puede ser explicada por las variables predictoras incluidas en el modelo. En resumen, el modelo muestra una capacidad significativa para prever el ingreso per cápita, lo que lo convierte en una herramienta valiosa para comprender y analizar los factores que influyen en el estándar de vida de diferentes países.
